```python
filtered_data = data[data["job_title"].isin(["Data Scientist", "Data Engineer", "Data Analyst"])]
filtered_data.head(20)
```

```python
sns.catplot(x='job_title', y='salary_in_usd', kind='box', data=filtered_data, height=7)
```

```python
plot_data = filtered_data.groupby(['job_title', 'experience_level'])['salary_in_usd'].mean().unstack()
plot_data
```

```python
plot_data = plot_data.reindex(index=['Data Analyst', 'Data Scientist', 'Data Engineer'])
plot_data = plot_data [['EN', 'MI', 'SE', 'EX']]
plot_data
```

```python
sns.catplot(x='job_title', y='salary_in_usd', hue='experience_level', data=filtered_data, height=6)
```

```python
plot_data.plot.bar(ylabel='USD salary')
```

```python
sns.lineplot(data=filtered_data, x='work_year', y='salary_in_usd', hue='job_title')
```
